```python

class Income:
    
    def __init__(self):
        self.income = 0
        self.expense = 0
        self.investments = 0

    def get_income(self):
        rent = int(input("What is your Monthly Rental income?: "))
        laundry = int(input("What is your Monthly Laundry income?: "))
        storage = int(input("What is your Monthly Storage income?: "))
        misc = int(input("What is your Monthly Miscellaneous income?: "))
        self.income = rent + laundry + storage + misc
    
    def get_expenses(self):
        tax = int(input("What is your Monthly Tax Expense?: "))
        insurance = int(input("What is your Monthly Insurance Expense?: "))
        utility = int(input("What is your Monthly Utility Expenses?\nThis would include Electric, Water, Sewer,\nGarbage, and Gas: ")) 
        hoa = int(input("What is your Monthly HOA Expense?: ")) 
        lawnsnow = int(input("What is your Monthly Lawn/Snow Expense?: ")) 
        mortgage = int(input("What is your Monthly Mortgage Payment?: ")) 
        vacancy = self.income * 0.05
        misc_expense = 500
        self.expense = self.income + tax + insurance + utility + hoa + lawnsnow + mortgage + vacancy + misc_expense

    def get_investments(self):
            down_pay = int(input("What is your Down Payment?: ")) 
            closing_cost = int(input("What is your Closing Costs?: "))
            rehab = int(input("What is your Rehab Budget?: ")) 
            misc_cost = int(input("What are your other costs?: ")) 
            self.investments = down_pay + closing_cost + rehab + misc_cost
    
    def calculate_roi(self):
        monthly_income = self.income - self.expense
        cash_flow = monthly_income * 12
        roi = cash_flow / self.investments
        print("---- Your Cash on Cash ROI Amount is: {0:.2f}% ----".format(roi))

def roi_calculator():
    user_1 = Income()
    go_again = 't'
    while go_again == 't':
        print('''
        Welcome to the ROI Calculator! Please follow the instructions below.
        Disclaimer: Please enter in Integers and no comma's, Round 
        your numbers up or down so the calculator can function!
        --------------------------------------------------------------------
        ''')
        user_1.get_income()
        user_1.get_expenses()
        user_1.get_investments()
        user_1.calculate_roi()
        go_again = input("Would you like to go again(Y/N)?: ")
        if go_again == 'N' or go_again == 'n' or go_again == 'No' or go_again == 'no':
            break
        if go_again == 'y' or go_again == 'Y' or go_again == 'Yes' or go_again == 'yes':
            roi_calculator()
        while go_again not in {'Y', 'y', 'N', 'n', 'Yes', 'yes', 'No', 'no'}:
            go_again = input("Your answer is invalid, please select Yes(Y) or No(N) to continue: ")
            if go_again == 'N' or go_again == 'n' or go_again == 'No' or go_again == 'no':
                break
            if go_again == 'y' or go_again == 'Y' or go_again == 'Yes' or go_again == 'yes':
                roi_calculator()
roi_calculator()        
        
```
